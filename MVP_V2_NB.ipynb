```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('backend\database\MVP_DB_V2_NY_CA.csv')
```

```python
df.head(2)
```

```python
userState = "NY"
userCountry = 'USA'
userCounty = "NEW YORK CITY"
```

```python
rebateData = df.copy()
fiteredRebateData = rebateData[ (rebateData["State"] == userCountry) | \
                                ((rebateData["State"] == userState) & (rebateData["County"] == userState)) | \
                                (rebateData["County"] == userCounty)]
```

```python
fiteredRebateData.head(2)
```

```python
fiteredRebateData['Estimated Incentive Value'] = fiteredRebateData['Estimated Incentive Value'].astype(float)
```

```python
rebates_data = fiteredRebateData[fiteredRebateData['Incentive Type'] == 'Discount']
tax_amount_data = fiteredRebateData[fiteredRebateData['Incentive Value'] != 'Discount']
```

```python
fiteredRebateData.columns
```

```python
resdf_rebate = rebates_data.groupby(['Technology' , 'Sub-Technology'])['Estimated Incentive Value'].agg(['mean'])
resdf_rebate
```

```python
resdf_tax = tax_amount_data.groupby(['Technology' , 'Sub-Technology'])['Estimated Incentive Value'].agg(['mean'])
resdf_tax
```

```python
resdf2 = fiteredRebateData.groupby(['Technology' , 'Sub-Technology','Incentive Name','Provider', 'Incentive Type'])[['Estimated Incentive Value']].agg(['mean'])
```

```python
resdf2 = rebates_data.groupby(['Technology' ,'Incentive Name','Provider'])[['Estimated Incentive Value']].agg(['mean']).reset_index()
resdf2[resdf2["Technology"] == 'Controls']
```
